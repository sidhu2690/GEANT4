```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
elec_df = pd.read_csv("/content/elec.csv")
muo_df = pd.read_csv("/content/muo.csv")
```

```python
electron_data = pd.read_csv("/content/electron_data.csv")
muon_data = pd.read_csv("/content/muon_data.csv")
```

```python
elec_df.head(5)
```

```python
electron_data.info()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

axes[0].scatter(elec_df['Layer'], elec_df['Mean_Energy'])
axes[0].set_title("Mean Energy Deposited vs Layer - (Electron Primary)")
axes[0].set_xlabel("Layer Number")
axes[0].set_ylabel("Mean Energy Deposited (MeV)")
axes[0].grid(True)

axes[0].set_xticks(np.arange(elec_df['Layer'].min(), elec_df['Layer'].max() + 1, 1))
axes[0].tick_params(axis='x', rotation=45, labelsize=9)

axes[0].text(
    0.3, 0.9,
    "• Mean energy deposition per layer\n"
    "• Electron primary particles",
    transform=axes[0].transAxes,
    fontsize=10,
    verticalalignment='top',
    bbox=dict(boxstyle="round,pad=0.4", facecolor='white', alpha=0.8)
)

axes[1].scatter(muo_df['Layer'], muo_df['Mean_Energy'])
axes[1].set_title("Mean Energy Deposited vs Layer - (Muons Primary)")
axes[1].set_xlabel("Layer Number")
axes[1].set_ylabel("Mean Energy Deposited (MeV)")
axes[1].grid(True)

axes[1].set_xticks(np.arange(muo_df['Layer'].min(), muo_df['Layer'].max() + 1, 1))
axes[1].tick_params(axis='x', rotation=45, labelsize=9)

axes[1].text(
    0.3, 0.9,
    "• Mean energy deposition per layer\n"
    "• Muon primary particles",
    transform=axes[1].transAxes,
    fontsize=10,
    verticalalignment='top',
    bbox=dict(boxstyle="round,pad=0.4", facecolor='white', alpha=0.8)
)

plt.tight_layout(pad=3.0)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

elec_sorted = elec_df.sort_values('Layer')
elec_cumsum = elec_sorted['Mean_Energy'].cumsum()

axes[0].plot(elec_sorted['Layer'], elec_cumsum, marker='o')
axes[0].set_title("Cumulative Energy Deposited vs Layer - (Electron Primary)")
axes[0].set_xlabel("Layer Number")
axes[0].set_ylabel("Cumulative Energy Deposited (MeV)")
axes[0].grid(True)

axes[0].set_xticks(np.arange(elec_sorted['Layer'].min(), elec_sorted['Layer'].max() + 1, 1))
axes[0].tick_params(axis='x', rotation=45, labelsize=9)

axes[0].text(
    0.3, 0.2,
    "• Cumulative sum of mean energy per layer\n"
    "• Electron primary particles",
    transform=axes[0].transAxes,
    fontsize=10,
    verticalalignment='top',
    bbox=dict(boxstyle="round,pad=0.4", facecolor='white', alpha=0.8)
)

muo_sorted = muo_df.sort_values('Layer')
muo_cumsum = muo_sorted['Mean_Energy'].cumsum()

axes[1].plot(muo_sorted['Layer'], muo_cumsum, marker='o')
axes[1].set_title("Cumulative Energy Deposited vs Layer - (Muon Primary)")
axes[1].set_xlabel("Layer Number")
axes[1].set_ylabel("Cumulative Energy Deposited (MeV)")
axes[1].grid(True)

axes[1].set_xticks(np.arange(muo_sorted['Layer'].min(), muo_sorted['Layer'].max() + 1, 1))
axes[1].tick_params(axis='x', rotation=45, labelsize=9)

axes[1].text(
    0.3, 0.2,
    "• Cumulative sum of mean energy per layer\n"
    "• Muon primary particles",
    transform=axes[1].transAxes,
    fontsize=10,
    verticalalignment='top',
    bbox=dict(boxstyle="round,pad=0.4", facecolor='white', alpha=0.8)
)

plt.tight_layout(pad=3.0)
plt.show()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(16, 6))

axes[0].scatter(electron_data['Layer'], electron_data['Raw_Hits'])
axes[0].set_title("Number of Raw Hits vs Layer - (Electron Primary)")
axes[0].set_xlabel("Layer Number")
axes[0].set_ylabel("Number of Raw Hits")
axes[0].grid(True)

axes[0].set_xticks(np.arange(electron_data['Layer'].min(), electron_data['Layer'].max() + 1, 1))
axes[0].tick_params(axis='x', rotation=45, labelsize=9)

axes[0].text(
    0.3, 0.9,
    "• Only hits with energy deposition > 0\n"
    "• Primary and secondary particles are not combined",
    transform=axes[0].transAxes,
    fontsize=10,
    verticalalignment='top',
    bbox=dict(boxstyle="round,pad=0.4", facecolor='white', alpha=0.8)
)

axes[1].scatter(electron_data['Layer'], electron_data['Filtered_Hits'])
axes[1].set_title("Number of Filtered Hits vs Layer - (Electron Primary)")
axes[1].set_xlabel("Layer Number")
axes[1].set_ylabel("Number of Filtered Hits")
axes[1].grid(True)

axes[1].set_xticks(np.arange(electron_data['Layer'].min(), electron_data['Layer'].max() + 1, 1))
axes[1].tick_params(axis='x', rotation=45, labelsize=9)

axes[1].text(
    0.3, 0.95,
    "• Only hits with energy deposition > 0\n"
    "• Primary and secondary particles are not combined\n"
    "• Hits below 25% MIP threshold removed",
    transform=axes[1].transAxes,
    fontsize=10,
    verticalalignment='top',
    bbox=dict(boxstyle="round,pad=0.4", facecolor='white', alpha=0.8)
)

plt.tight_layout(pad=3.0)
plt.show()
```
