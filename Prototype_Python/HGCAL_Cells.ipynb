```python
cellsize = 7  # mm
inner_radius = 150  #mm
outer_radius = [150, 142, 150, 142, 150, 142, 150, 142, 150, 150,
        150, 150, 150, 150, 150, 150, 150, 158, 158, 150,
        158, 150, 158, 158, 158, 158, 158, 164, 167, 169,
        167, 178, 183, 197, 206, 215, 215, 234, 245, 256,
        256, 256, 256, 256, 256, 256, 245] #cm

num_cells = []

for i in outer_radius:
  max_index = int(round(i*10 / cellsize ))
  min_index = -max_index

  cell_count = 0

  for j in range(min_index, max_index + 1):
    for k in range(min_index, max_index + 1):
      xi = j * cellsize
      yi = k * cellsize
      r  = (xi**2 + yi**2) ** (1/2)

      if inner_radius <= r <= i*10:
        cell_count += 1
  num_cells.append(cell_count)
```

```python
sum(num_cells)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (6, 4))

plt.plot([i for i in range(1, 48)],num_cells )
plt.ylabel("Number of cells")
plt.xlabel("Layer")
plt.show()
```
